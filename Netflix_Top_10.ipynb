## Fill In Missing Values

Mengisi beberapa missing values #Nan
dan mengganti data tersebut dengan value `0`

```python
# fill some null values
def fill_null():
    data = {'Rating':0.0,'Imdb Score':'0', 'Duration' : 0, 'Release Date' : 0, 'Duration' : 0}
    for col,val in data.items():
        print(col)
        df[col] = df[col].fillna(val)
    return df
data_to_clean = fill_null()
data_to_clean.info()
data_to_clean.head()
```

```python
mean_country = df.groupby(['Production Country']).mean().reset_index()
mean_country
```

```python
# fill some null values
def fill_null():
    data = {'Production Country' : 'mean_country'}
    for col,val in data.items():
        print(col)
        df[col] = df[col].fillna(val)
    return df
data_to_clean = fill_null()
data_to_clean.info()
data_to_clean.head()
```
